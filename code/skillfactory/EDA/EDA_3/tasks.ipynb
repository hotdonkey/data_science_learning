```python
import pandas as pd
import statistics
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/hotdonkey/Documents/GitHub/data_science_learning/code/skillfactory/EDA/data/wine_cleared.csv')
```

```python
data['price_round'] = data['price'].round().astype(int)
```

```python
regex = '\d{4}' # регулярное выражение для нахождения чисел
data['year'] = data['title'].str.findall(regex).str.get(0)
```

```python
data.head()
```

```python
data['is_usa'] = data['country'].apply(lambda x: 1 if x == 'US' else 0)
```

```python
#Task 2.1
data['country'].value_counts(sort=True,ascending=False).head()
```

```python
data['is_france'] = data['country'].apply(lambda x: 1 if x == 'France' else 0)
data['is_italy'] = data['country'].apply(lambda x: 1 if x == 'Italy' else 0)
```

```python
#Task 2.2
print(data['is_france'].sum(),data['is_italy'].sum())
```

```python
#Task 2.3
data['year'] = data['year'].apply(pd.to_numeric)
data['old_wine'] = data['year'].apply(lambda x:  1 if x < 2010 else 0)
data['old_wine'].sum()
```

```python
regex_2 = '\(+.{1,}'
data['locality'] = data['title'].str.findall(regex_2).str.get(0)
data['locality'] = data['locality'].str.replace('(','').str.replace(')','')
```
